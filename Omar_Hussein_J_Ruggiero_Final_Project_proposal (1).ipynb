## Part I
The first data source of this project is a CSV from Kaggle [5]. This dataset contains a list of video games with sales greater than 100,000 copies and it was generated by a scrape of vgchartz.com based on BeautifulSoup using Python.

The vgsales fields include:

- Rank - Ranking of overall sales
- Name - The games name
- Platform - Platform of the games's release (i.e. PC, PS4, etc.)
- Year - Year of the game's release
- Genre - Genre of the game
- Publisher - Publisher of the game
- NA_Sales - Sales in North America (in millions)
- EU_Sales - Sales in Europe (in millions)
- JP_Sales - Sales in Japan (in millions)
- Other_Sales - Sales in the rest of the world (in millions)
- Global_Sales - Total worldwide sales.

### Algorithm Description
1. Pre-process each text (i.e. remove HTML tags, noisy characters). 
2. Initialize the total text sentiment score: s ← 0.
3. Tokenize text. For each token, check if it is present in a sentiment dictionary.
 - (a) If token is present in dictionary,
 - i. If token is positive, then s ← s + w.
 - ii. If token is negative, then s ← s − w.
4. Look at total text sentiment score s,
 - (a) If s > threshold, then classify the text as positive.
 - (b) If s < threshold, then classify the text as negative.

In dictionary-based techniques the idea is to first collect a small set of opinion words manually with
known orientations, and then to grow this set by searching in the WordNet dictionary for their synonyms and antonyms which we will find using NLTK library. The newly found words are added to the seed list. The next iteration starts. The iterative process stops when no more new words are
found [6].

The final result would include the construction of a choropleth map that will show how positive and negative tweets distribution varies between contries, or at a state level.
